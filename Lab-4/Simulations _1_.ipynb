Simulation¶  
Simulation is the process of using a computer to mimic a physical experiment. Those experiments will almost invariably involve chance.We have seen how to simulate the results of casting a dice. The steps in that simulation
were examples of the steps that will constitute every simulation we do in this course. 

Step 1: What to Simulate¶
Specify the quantity you want to simulate. For example, you might decide that you want to simulate the outcomes of tosses of a coin.

Step 2: Simulating One Value¶
Figure out how to simulate one value of the quantity you specified in Step 1. Eg - figure out how to simulate the outcome of one toss of a coin/ dice. If your quantity is more complicated, you might need several lines of code to come up with one simulated value.

Step 3: Number of Repetitions¶
Decide how many times you want to simulate the quantity. You will have to repeat Step 2 that many times. Eg 10000 tosses of a coin, and so we toss 10000 times the step of generating the outcome of a single toss.

Step 4: Coding the Simulation¶
Put it all together in code.
Create an empty array in which to collect all the simulated values. Create a "repetitions sequence," that is, a sequence whose length is the number of repetitions you specified in Step 3. For n repetitions we will almost always use the
sequence np.arange(n). Create a for loop. For each element of the repetitions sequence: Simulate one value based on the code you developed in Step 2.Augment the collection array with this simulated value.

```python
import numpy as np
```

```python
sides = ["Heads", "Tails"]
coin = np.array(sides)
```

```python
coin
```

```python
outcomes = np.array
```

```python
for i in np.arange(5):
    outcome_of_toss = np.random.choice(coin)
outcome_of_toss
```

```python
for i in np.arange(5):
    outcome_of_toss = np.random.choice(coin)
    #outcomeA = outcome_of_toss
    outcomes = np.append(outcomes, outcome_of_toss)
outcomes
```

By capturing the results in an array we have given ourselves the ability to use array methods to do computations. For example, we can use np.count_nonzero to count the number of heads in the five tosses.

```python
np.count_nonzero(outcomes == 'Heads')
```

Iteration is a powerful technique. For example, by running exactly the same code for 100 tosses instead of 5, we can count the number of heads in 100 tosses

```python
for i in np.arange(100):
    outcome_of_toss = np.random.choice(coin)
    outcomes = np.append(outcomes, outcome_of_toss)
np.count_nonzero(outcomes == 'Heads')
```

Step 3: Number of Repetitions¶
How many repetitions we want is up to us. The more we use, the more reliable our simulations will be, but the longer it will take to run the code. Let's go for 10,000 repetitions. That means we are going to do the following 10,000
times: Toss a coin 100 times and count the number of heads.

```python
#Step 4: Coding the Simulation
```

```python
# Repetitions sequence
num_repetitions = 10000
```

```python
outcome = np.random.choice(coin, 100)
outcome
```

```python
num_heads = np.count_nonzero(outcome == 'Heads')
num_heads
```

```python
# An empty array to collect the simulated values
heads = np.array
```

```python
repetitions_sequence = np.arange(num_repetitions)
repetitions_sequence
```

```python
# for i in repetitions_sequence:
# # simulate one value
#     outcome = np.random.choice(coin, 100)
#     num_heads = np.count_nonzero(outcomes == 'Heads')
#     # augment the collection array with the simulated value
#     heads = np.append(heads, num_heads)
```

```python
repetitions_sequence = np.arange(num_repetitions)
# for loop

for _ in repetitions_sequence:
# simulate one value
    outcome = np.random.choice(coin, 100)
    num_heads = np.count_nonzero(outcome == 'Heads')
    print(num_heads)
    # augment the collection array with the simulated value
    heads = np.append(heads, num_heads)
# That's it
```
