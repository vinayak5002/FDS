### Part A

#### Null hypothesis
 - The percentage of purple flowers is 75%  
 => mu = 75%
#### Alternate hypothesis
 - The percentage of purple flowers is not 75%  
 => mu != 75%

#### We can use normal 2 tailed hypothesis testing for this case

#### Observed Statistic

Observed no. of purple flowers = 705 + 65 = 770 (CB.EN.U4CES20465)

Total number of flowers = 929

observed percentage = (770/929)*100 = 82%

Therefore, Observed Statistic = 82.88%

```python
import numpy as np
```

```python
flowers = np.random.choice(2, 5, p=[0.75, 0.25]) # Generates  20 number of 0 or 1 with probabilites 0.75 and 0.25
print(flowers)
np.count_nonzero(flowers == 0)
```

Similating 200-465 times = 1535 times
Size of sample is 705+65 = 770

```python
percentages = []

for i in range(1535):
    flowers = np.random.choice(2, 929, p=[0.75, 0.25])
    no_purple_flowers = np.count_nonzero(flowers == 0)
    percent_purple = (no_purple_flowers/929)*100
    percentages.append(percent_purple)

percentages
```

```python
np.mean(percentages)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(percentages)
plt.scatter([82.88], [1], color='red')
plt.show()
```

Since our observed statistic is not even close to our sample mean, we reject null hypothesis

### Part B

```python
import scipy.stats as st
```

### Hypothesis : 

Null hypothesis :
 - mu = 60  

Alternate hypothesis:
 - mu != 60

```python
# Given

std = 13.46
mu = 60
n = 30
x_bar = 65.10
alpha = 0.05
```

```python
# test statistic

t = (x_bar - mu)/(std/np.sqrt(30))
t
```

```python
# Z_critical value for alpha = 0.05

z_critical = abs(st.norm.ppf(alpha/2))
z_critical
```

```python
if(t < z_critical):
    print("Failed to reject null hypothesis")
else:
    print("Reject null hypothesis")
```

Therefore, mean of the population is not 60
