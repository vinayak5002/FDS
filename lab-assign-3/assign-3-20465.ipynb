```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./train.csv')
```

```python
from sklearn.impute import SimpleImputer
imp = SimpleImputer(missing_values= np.nan, strategy= 'mean')

df.loc[:, ['Age']]= imp.fit_transform(df[['Age']])
```

### Histograms

```python
plt.hist(df['Age'], bins=range(0, 101, 5))

plt.show()
```

### Bar chart

```python
embarked = dict(df['Embarked'].value_counts())
```

```python
embarked
```

```python
plt.bar(embarked.keys(), embarked.values())
plt.show()
```

### Box Plots

```python
plt.boxplot(df['Age'])

plt.show()
```

### Scatter plot

```python
bost = pd.read_csv('./data.csv')
```

### Pie charts

```python
plt.pie(embarked.values(), labels=embarked.keys())
plt.show()
```

### Mean, median, mode

```python
fare = df['Fare']

print(f'Mean: {fare.mean()}\nMedian: {fare.median()}\nMode: {fare.mode()}')
```

### Variance, Std deviation, Covariance, Correlation covefficient

```python
dis = {
    "Variance" : fare.var(),
    "Standard deviation": fare.std(),
    "Covariance" : fare.cov(df['Age']),
    "Correlation coefficient": fare.corr(df['Age'])
}
dis
```
